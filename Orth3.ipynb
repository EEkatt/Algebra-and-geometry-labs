**(17 баллов)** По заданным 

- симметрической матрице $A$ размера $3\times 3$
- вектор-строке $b$ длины $3$
- коэффициенту $c_0$

вывести тип поверхности второго порядка, заданной уравнением
$$x^T A x + 2b x + c_0 = 0.$$

Кроме этого, вывести каноническую ортогональную систему координат (начало координат и базис).

```python

import numpy as np
import scipy.linalg as sla


def I_1 (A):
    i_1 = 0
    for i in range (3):
        i_1 += A[i][i]
    return i_1


def I_2 (A):
    i_2 = 0
    for i in range (2):
        for j in range(1, 3):
            if i != j:
                i_2 += sla.det([[A[i][i], A[i][j]], [A[i][j], A[j][j]]])
    return i_2


def I_4 (A, b, c):
    i_4 = A
    i_4 =  np.vstack([i_4, b])
    b_c = np.insert(b, 3, c)
    i_4 = np.hstack([i_4, np.atleast_2d(b_c).T])
    return i_4


def K_2 (A, b, c):
    i_4 = I_4(A, b, c)
    k_2 = 0
    for i in range(3):
        k_2 += sla.det([[i_4[i][i], i_4[i][3]], [i_4[i][3], i_4[3][3]]])
    return k_2


def K_3 (A, b, c):
    i_4 = I_4(A, b, c)
    k_3 = 0
    for i in range(3):
        for j in range (3):
            if i < j:
                k_3 += sla.det([[i_4[i][i], i_4[i][j], i_4[i][3]], [i_4[i][j], i_4[j][j], i_4[j][3]], [i_4[i][3], i_4[j][3], i_4[3][3]]])
    return k_3


def type (A, b, c):
    A = A.astype(np.float64)
    b = b.astype(np.float64)
    det_A = sla.det(A)
    if det_A != 0:
        if I_2(A) > 0 and I_1(A)*det_A > 0:
            if sla.det(I_4(A, b, c)) > 0:
                print("Мнимый эллипсоид")
            elif sla.det(I_4(A, b, c)) < 0:
                print("Эллипсоид")
            else:
                print("Мнимый конус")
        elif I_2(A) <= 0 or I_1(A)*det_A <=0:
            if sla.det(I_4(A, b, c)) > 0:
                print("Двуполостный гипербалоид")
            elif sla.det(I_4(A, b, c)) < 0:
                print("Однополостный гипербалоид")
            else:
                print("Конус")
    else:
        if sla.det(I_4(A, b, c)) > 0:
            print("Гиперболтчесикй параболоид")
        elif sla.det(I_4(A, b, c)) < 0:
            print("Эллиптическй параболоид")
        else:
            if I_2(A) > 0:
                if I_1(A)*K_3(A, b, c) > 0:
                    print("МИнимый эллиптическй цилиндр")
                elif I_1(A)*K_3(A, b, c) < 0:
                    print("Эллиптическй цилиндр")
                elif K_3(A, b, c) == 0:
                    print("Пара мнимых пересекающихся плоскостей")
            elif I_2(A) < 0:
                if K_3(A, b, c) != 0:
                    print("Гиперболический цилиндр")
                else:
                    print("Пара пересекающихся плоскостей")
            else:
                if K_3(A, b, c) != 0:
                    print("Параболический цилиндр")
                else:
                    if K_2(A, b, c) < 0:
                        print("Пара параллельных плоскостей")
                    elif K_2(A, b, c) > 0:
                        print("Пара мнимых параллельных плоскостей")
                    else:
                        print("Пара совпадающих плоскостей")

if __name__ == '__main__':
    A = np.array([[7, -2, 0], [-2, 6, -2], [0, -2, 5]])
    b = np.array([-3, -12, 9])
    c = 30
    type(A, b, c)
    
    A = np.array([[1, 0, 1], [0, 2, 0], [1, 0, 3]])
    b = np.array([1, 0, -3])
    c = 10
    type(A, b, c)
    
    A = np.array([[2, 2, 1], [2, 2, 1], [1, 1, 3]])
    b = np.array([-2, 3, -1])
    c = 3
    type(A, b, c)
    
    A = np.array([[2, -2.5, 1], [-2.5, 2, 1], [1, 1, -4]])
    b = np.array([-4, -4, 2.5])
    c = -1
    type(A, b, c)
```
